### Retail Trader Activity

```python
# get historic values, process data
df = pd.read_csv('NDAQ_RTAT_12_5_22.csv')
df.dropna(inplace=True)
```

### Feature Creation

```python
#ticker based window metrics
window = df.set_index('date').sort_index().groupby('ticker').rolling(3).mean()
```

```python
window[window.index.get_level_values(0) =='FSLR']
```

```python
#daily market wide metrics
avgsent = df.groupby('date').mean()['sentiment']
stdsent = df.groupby('date').std()['sentiment']
stdact = df.groupby('date').std()['activity']
```

```python
# sector/industry market wide metrics (after getting equities bundle)
```

```python
#create a metric that incorporates sentiment and activity
#right now it's more driven by activity
# activity itself seems like it could be a good predictor of price drop
df['product_sentiment_activity'] = (df['activity']+.00001)*df['sentiment']*100
```

```python
df[df['date']>'2021-01-01'].sort_values('product_sentiment_activity',ascending=False).head(20)
```
